# Ejercicios manejo básico de pandas

Importamos la libreria pandas

```python
import pandas as pd
```

Leemos un archivo .csv desde el directorio raíz

```python
archivo=pd.read_csv('rpi_data_long.csv')
archivo
```

Creamos un arreglo para reemplazar los nombres de las columnas del archivo original

```python
column_names = [ 'Tipo A', 'Medida A', 'Unidad A',
                 'Tipo B', 'Medida B', 'Unidad B',
                 'Tipo C', 'Medida C', 'Unidad C', 
                 'Fecha']
```

Reeemplazamos los nombres de las columnas

```python
archivo.columns = column_names
archivo
```

Copiamos el archivo enun archivo auxiliar

```python
auxiliar= archivo.copy()
auxiliar
```

Cambiamos el nombre de las columnas en relación con las medidas como se muestra:

    Medida A -> Ping (ms)
    Medida B -> Descargar (Mbit/s)
    Medida C -> Cargar (Mbit/s)

```python
auxiliar.rename(columns={'Medida A':'Ping (ms)', 
                           'Medida B': 'Download (Mbit/s)',
                           'Medida C': 'Upload (Mbit/s)'}, inplace=True)
auxiliar.head(3)
```

Eliminaremos las columnas de los tipos

```python
auxiliar.drop(['Tipo A', 'Tipo B', 'Tipo C'], axis=1, inplace=True)
auxiliar.head()
```

Elimine las columnas de las Unidades

```python
# Code cell 28
auxiliar['Día'] = auxiliar['Fecha'].apply(lambda x: pd.to_datetime(x).date())
```
